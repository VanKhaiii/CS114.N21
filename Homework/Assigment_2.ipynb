# DATA VISULIZATION

```python
import os
import cv2
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

dataset_folder = '/content/dataset'
subfolders = [os.path.join(dataset_folder, str(i)) for i in range(10)]

image_paths = []
labels = []

for subfolder in subfolders:
    label = int(os.path.basename(subfolder))
    #print(f"Loading data in floder {label}")
    subfolder_images = [os.path.join(subfolder, image) for image in os.listdir(subfolder)]
    image_paths.extend(subfolder_images)
    labels.extend([label] * len(subfolder_images))

images, labels = shuffle(image_paths, labels, random_state=42)
```

# SVM

### Feature extraction 1: Resize image to 28 x 28 -> flatten

```python
from tqdm import tqdm
def extract_features(img_file):
    img = cv2.imread(img_file, cv2.IMREAD_GRAYSCALE)
    resized_img = cv2.resize(img, (8, 8), interpolation=cv2.INTER_AREA)
    return resized_img.reshape((64))

def extract_features_from_dataset(img_paths):
  X = []
  for path in tqdm(img_paths):
    X.append(extract_features(path))
  return np.array(X)

X_train = extract_features_from_dataset(images_train)
X_test = extract_features_from_dataset(images_test)
```

```python
from sklearn import datasets, svm, metrics

# Create a classifier: a support vector classifier
clf = svm.SVC(gamma=0.00001)

# Learn the digits on the train subset
clf.fit(X_train, y_train)

# Predict the value of the digit on the test subset
predicted = clf.predict(X_test)
```

```python
_, axes = plt.subplots(nrows=1, ncols=4, figsize=(10, 3))
for ax, image, prediction in zip(axes, X_test, predicted):
    ax.set_axis_off()
    image = image.reshape(8, 8)
    ax.imshow(image, cmap='gray', interpolation="nearest")
    ax.set_title(f"Prediction: {prediction}")
```

```python
print(
    f"Classification report for classifier {clf}:\n"
    f"{metrics.classification_report(y_test, predicted)}\n"
)

disp = metrics.ConfusionMatrixDisplay.from_predictions(y_test, predicted)
disp.figure_.suptitle("Confusion Matrix")
print(f"Confusion matrix:\n{disp.confusion_matrix}")
```

### Feature extraction 3: Threshold image -> Crop image -> Resize image to 28 x 28 -> flatten

```python
def crop_bbox(bin_img):
  contours,hierarchy = cv2.findContours(bin_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
  if not contours:
    return False, bin_img
  c = max(contours, key = cv2.contourArea)
  x,y,w,h = cv2.boundingRect(c)
  return True, bin_img[y: y+h, x:x+w]

def extract_features(img_file):
    img = cv2.imread(img_file, cv2.IMREAD_GRAYSCALE)
    ret, thresh = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY_INV)
    ret, cropped_img = crop_bbox(thresh)
    if not ret:
        print(f'failed to crop {img_file}')
        resized_img = cv2.resize(img, (8, 8), interpolation=cv2.INTER_AREA)
    else:
        resized_img = cv2.resize(cropped_img, (8, 8), interpolation=cv2.INTER_AREA)

    resized_img = resized_img.reshape((64))
    resized_img = resized_img
    return resized_img

def extract_features_from_dataset(img_paths):
  X = []
  for path in tqdm(img_paths):
    X.append(extract_features(path))
  return np.array(X)

X_train = extract_features_from_dataset(images_train)
X_test = extract_features_from_dataset(images_test)
```

# CNN (Nguồn: https://www.kaggle.com/code/arunrk7/digit-recognition-using-cnn-99-accuracy)

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D
from keras.optimizers import Adam
from keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint
import keras
```

### Data preparation

```python
def extract_features_CNN(img_file):
    img = cv2.imread(img_file, cv2.IMREAD_GRAYSCALE)
    resized_img = cv2.resize(img, (28, 28), interpolation=cv2.INTER_AREA)
    resized_img = resized_img / 255
    return resized_img

def extract_features_from_dataset_CNN(img_paths):
  X = []
  for path in tqdm(img_paths):
    X.append(extract_features_CNN(path))
  return np.array(X)
```

```python
# Read gray-scale images and resize to size 28 x 28
X_train = extract_features_from_dataset_CNN(images_train)
X_val = extract_features_from_dataset_CNN(images_val)
X_test = extract_features_from_dataset_CNN(images_test)

# Convert label to ont hot encoding
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False)
y_train = encoder.fit_transform(np.array(y_train).reshape(-1, 1))
y_val = encoder.fit_transform(np.array(y_val).reshape(-1, 1))
y_test = encoder.fit_transform(np.array(y_test).reshape(-1, 1))

print(f'\n Train: {X_train.shape[0]}\n Val: {X_val.shape[0]}\n Test: {X_test.shape[0]}')
```

### Build CNN architecture

```python
model = Sequential()

model.add(Conv2D(filters = 32, kernel_size = (5,5),padding = 'Same',
                 activation ='relu', input_shape = (28,28,1)))
model.add(Conv2D(filters = 32, kernel_size = (5,5),padding = 'Same',
                 activation ='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))


model.add(Conv2D(filters = 64, kernel_size = (3,3),padding = 'Same',
                 activation ='relu'))
model.add(Conv2D(filters = 64, kernel_size = (3,3),padding = 'Same',
                 activation ='relu'))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
model.add(Dropout(0.25))


model.add(Flatten())
model.add(Dense(256, activation = "relu"))
model.add(Dropout(0.3))
model.add(Dense(10, activation = "softmax"))

model.summary()
```

```python
from keras import backend as K

optimizer = Adam(0.001)

model.compile(optimizer = optimizer , loss = "categorical_crossentropy", metrics=["accuracy"])

es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=20, min_delta=1) #, best_only=True

mc = ModelCheckpoint('best_model.h5', monitor='val_accuracy', mode='max', verbose=1)

batch_size=95
epoch=50
```

```python
history = model.fit(X_train,y_train, batch_size=batch_size,
                    epochs = epoch, validation_data = (X_val,y_val),
                    callbacks=[es, mc])
```

```python
plt.figure()
fig,(ax1, ax2)=plt.subplots(1,2,figsize=(10,4))
ax1.plot(history.history['loss'])
ax1.plot(history.history['val_loss'])
ax1.legend(['training','validation'])
ax1.set_title('loss')
ax1.set_xlabel('epoch')

ax2.plot(history.history['accuracy'])
ax2.plot(history.history['val_accuracy'])
ax2.legend(['training','validation'])
ax2.set_title('Acurracy')
ax2.set_xlabel('epoch')



score =model.evaluate(X_test,y_test,verbose=0)
print('Test Score:',score[0])
print('Test Accuracy:',score[1])
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from keras.models import load_model

model = load_model('best_model.h5')

pred = model.predict(X_test)
# Convert predictions classes to one hot vectors
pred_classes = np.argmax(pred,axis = 1)
# Convert validation observations to one hot vectors
y_test_true = np.argmax(y_test,axis = 1)
print(f"Classification report for classifier {model}:\n"
      f"{metrics.classification_report(y_test_true, pred_classes)}\n")
# compute the confusion matrix
confusion_matrix = metrics.confusion_matrix(y_test_true, pred_classes)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix)

cm_display.plot()
plt.show()
```
